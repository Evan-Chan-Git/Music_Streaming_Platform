```python
# Query 1:  Simple Query
# (Find songs where 'genre' is 'pop')

# Access your collection 
collection = db['song']

# Define the query
query = {'genre': 'pop'}

try:
    # Execute the query
    results = collection.find(query)

    # Print the results
    for result in results:
        print(result)

except Exception as e:
    print(f"Error executing query: {e}")
```

```python
# Query 2:  Complex Query
# (Find the creators of each playlist )

# Access your collection 
collection = db['creates']

# Define the aggregation pipeline
pipeline = [
    # Join creates_collection with user_collection on user_id
    {"$lookup": {
        "from": "user",
        "localField": "user_id",
        "foreignField": "user_id",
        "as": "user"
    }},
    # Unwind the user array (length 1 due to join)
    {"$unwind": "$user"},
    # Join with playlist_collection on playlist_id
    {"$lookup": {
        "from": "playlist",
        "localField": "playlist_id",
        "foreignField": "playlist_id",
        "as": "playlist"
    }},
    # Unwind the playlist array
    {"$unwind": "$playlist"},
    # Project to shape the output
    {"$project": {
        "_id": 0,
        "userName": "$user.name",
        "playlistName": "$playlist.name"
    }}
]

try: 
    # Execute the aggregation pipeline
    result = list(collection.aggregate(pipeline))

    # Display the result
    for doc in result:
        print(doc)

except Exception as e:
    print(f"Error executing query: {e}")
```

```python
# Query 3:  Aggregate/MapReduce Query (Pymongo version: 4.6.3 does not directly support MapReduce - depracated)
# (Find the count of users by gender)

# Access your collection 
collection = db['user']

# Define the aggregation pipeline
pipeline = [
    {"$group": {"_id": "$gender", "count": {"$sum": 1}}}
]

try:   
    # Execute the aggregation pipeline
    result = collection.aggregate(pipeline)

    # Print the result
    for doc in result:
        print(doc)

except Exception as e:
    print(f"Error executing query: {e}")
```
